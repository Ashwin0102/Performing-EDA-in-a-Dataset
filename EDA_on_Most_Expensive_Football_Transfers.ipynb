```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
#dataset
df=pd.read_csv('/content/football_players.csv')
df.index.name='S.No'
df.head()
```

```python
df['Position'].value_counts()
```

```python
df1=df.groupby(by='From(Club)',as_index=False, sort=True)['Fee(€ mln)'].sum()
df1=df1.sort_values('Fee(€ mln)',ascending=False)
df1.head()
```

```python
plt.figure(figsize=(50,30))
plt.xlabel('From(Club)')
plt.ylabel('Fee(€ mln)')
plt.title('Clubs in the order of highest transfer revenues')
plt.bar(df1['From(Club)'],df1['Fee(€ mln)'])
```

```python
df2=df.groupby(by='To(Club)',as_index=False, sort=True)['Fee(€ mln)'].sum()
df2=df2.sort_values('Fee(€ mln)',ascending=False)
df2.head()
```

```python
plt.figure(figsize=(25,15))
plt.xlabel('From(Club)')
plt.ylabel('Fee(€ mln)')
plt.title('Clubs in the order of highest transfer revenues')
plt.bar(df2['To(Club)'],df2['Fee(€ mln)'])
```

```python
df3=df.groupby('Fee(€ mln)',as_index=False)['Year'].median()
df3.head()
```

```python
plt.figure(figsize=(20,3))
plt.xlabel('Fee(€ mln)')
plt.ylabel('Year')
plt.title('Fees vs Year')
plt.plot(df3['Fee(€ mln)'],df3['Year'])
```

```python
df4=df.groupby(by='From(Country)',as_index=False)['Fee(€ mln)'].sum()
df4=df4.sort_values(by='Fee(€ mln)',ascending=False)
df4.head()
```

```python
plt.title('Money spent to buy players from different countries')
plt.xlabel('From(Countries)')
plt.ylabel('Fee(€ mln)')
plt.bar(df4['From(Country)'],df4['Fee(€ mln)'])
```

```python
df5=df.groupby(by='To(Country)',as_index=False)['Fee(€ mln)'].sum()
df5=df5.sort_values(by='Fee(€ mln)',ascending=False)
df5.head()
```

```python
plt.title('Money spent on buying players for different countries')
plt.xlabel('To(Countries)')
plt.ylabel('Fee(€ mln)')
plt.bar(df5['To(Country)'],df5['Fee(€ mln)'])
```

```python
df6=df.groupby(by='Position',as_index=False)['Fee(€ mln)'].mean()
df6.head()
```

```python
plt.bar(df6['Position'],df6['Fee(€ mln)'])
plt.xlabel('Position')
plt.ylabel('Fee(€ mln)')
plt.title('Average salary according to the position')
```
